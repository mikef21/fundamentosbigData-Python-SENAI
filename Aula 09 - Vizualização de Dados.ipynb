```python
### MULTIPLAS CURVAS NO MESMO MECANISMO GRÁFICO
import numpy as np    

x= np.linspace(0,2*np.pi,100)
y= np. sin(x)
z= np.cos(x)

plt.plot(x,y,label ='FUNÇÃO SENO',c= '#FF19B0')
plt.plot(x,z, label= 'FUNÇÃO COSSENO', color='#637BFF')

#ADICIONAR LEGENDA AO GRÁFICO 
plt.legend()

#ADICIONAR TITULO AO GRÁFICO 
plt.title('Funções Trigonométricas')

#ADICIONAR TITULO AOS EIXOS

plt.xlabel('Valores de X')

plt.ylabel('Valores de Y')

#SALVAR O GRÁFICO COMO IMAGEM 
plt.savefig('./Gráficos/funcoes-trigonometricas.png')
```

### EXEMPLO 2

Escreva um programa Python para desenhar um gráfico de dispersão para encontrar o aumento do nível do mar nos últimos 100 anos.
Esses dados contêm "mudanças cumulativas no nível do mar para os oceanos do mundo desde 1880, com base em uma combinação de medições de marés de longo
prazo e medições recentes de satélite. Eles mostram a mudança média absoluta do nível do mar, que se refere à altura da superfície do oceano, independentemente de
a terra próxima estar subindo ou descendo. Os dados de satélite baseiam-se exclusivamente no nível do mar medido, enquanto os dados do marégrafo de longo
prazo incluem um pequeno fator de correção porque o tamanho e a forma dos oceanos estão mudando lentamente ao longo do tempo. (Em média , o fundo do
oceano tem afundado gradualmente desde o último pico da Idade do Gelo, há 20.000 anos.)"

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

caminhoArquivo= ('./Arquivos/sea-level.csv')

df = pd.read_csv(caminhoArquivo)

#Fatiar o dataframe (selecionar a partir da linha 5 (5:) e todas as colunas(:))

df= df.loc[6:,:]
df.columns =[
             'Year',
             'CSIRO - Adjusted sea level (inches)',
             'CSIRO - Lower error bound (inches)',
             'CSIRO - Upper error bound (inches)',
             'NOAA - Adjusted sea level (inches)']

df.head()
# Converter os dados em numeros

df['Year'] = df['Year'].astype(int)
df['CSIRO - Adjusted sea level (inches)'] = df['CSIRO - Adjusted sea level (inches)'].astype(float)
df['CSIRO - Lower error bound (inches)'] = df['CSIRO - Lower error bound (inches)'].astype(float)
df['CSIRO - Upper error bound (inches)'] = df['CSIRO - Upper error bound (inches)'].astype(float)
df['NOAA - Adjusted sea level (inches)'] = df['NOAA - Adjusted sea level (inches)'].astype(float)

anos = df['Year'].to_list()
nivel = df['CSIRO - Adjusted sea level (inches)'].to_list()



plt.title('GRAFICO NIVEL DO MAR')

sns.scatterplot(x= anos, y = nivel)

plt.xlabel('Year')

plt.ylabel('CSIRO - Adjusted sea level (inches)')
```
