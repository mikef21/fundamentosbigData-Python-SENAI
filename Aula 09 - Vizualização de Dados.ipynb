```python

# Importando as bibliotecas 

import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np

import pandas as pd
```

### Primeiros passsos com Matplotlib

```python
# Criando os arrays (matrizes) para utilizar os gráficos


x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criando um gráfico com os valores anteriores

plt.plot(x,y)

```

### MODIFICAR O TIPO DE MARCADOR


```python
plt.plot(x,y, marker  = 'X', c= 'green')
```

```python
plt.plot(
    x,
    y,
    marker = "*",
    c='blue',
    ds = 'steps', # Forma de ligação
    ls ='--' #Estilo da Linha
)
```

```python
plt.plot(
    x, # Vetor dos valores do eixo X
    y, #  Vetor dos valores do eixo Y
    marker = "X",
    c='red', # Cor da linha
    ds = 'steps', # Forma de ligação
    ls ='--', #Estilo da Linha
    mec = 'blue' # Cor do marcador
)
```

### Adicionando outros elementos ao gráfico

```python
plt.plot(x,y) # Criando o gráfico

# Adicionando o título do eixo X e Y

plt.xlabel ("Valores de X", fontsize = 12)
plt.ylabel ("Valores de Y", fontsize = 12)

# Adicionando o titulo do gráfico

plt.title('Titulo do Gráfico',fontsize = 20)

# Adicionando as cores do rótulo de x e y

plt.xticks(fontsize = 12, c = 'blue')
plt.yticks(fontsize = 12, c = 'purple')
```
