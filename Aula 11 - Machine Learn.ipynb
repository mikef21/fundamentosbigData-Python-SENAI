### SCIKIT LEARN - DATA SETS EMBUTIDOS


```python
### Carregar o data set iris

from sklearn.datasets import load_iris

iris = load_iris(return_X_y= True)

len(iris)
```

### Separar os dados entre Treino e Teste

```python
## Importar a biblioteca para separar os dados

from sklearn.model_selection import train_test_split

## Separar os dados de x e y

X,y = iris 

print(len (X))

print(len(y))

## Fatiar os dados em Treino e teste

X_train, X_test, y_train, y_test = train_test_split (X,y, test_size= 0.3)


```

```python
#Tamanho dos dados (amostras) de treino
print('Dados de X para treinamento', len (X_train))
print('Dados de y para treinamento', len (y_train))

# Tamanho dos dados (amostras) para teste 
print('Dados de X para teste ', len(X_test))
print('Dados de y para teste ', len(y_test))
```

### Exercício 01 – Treino e Teste



```python

### Carregar o data set iris

from sklearn.datasets import load_diabetes

diabetes = load_diabetes(return_X_y= True)

len(diabetes)

```

```python
## Importar a biblioteca para separar os dados

from sklearn.model_selection import train_test_split

## Separar os dados de x e y

X,y = diabetes

print(len (X))

print(len(y))

## Dividindo a amostra 

X_train, X_test, y_train, y_test = train_test_split (X,y, test_size= 0.35, random_state=42)

print('Treino: %d' %len(X_train))
print('Teste: %d' %len(X_test))


#Tamanho dos dados (amostras) de treino
print('Dados de X para treinamento', len (X_train))
print('Dados de y para treinamento', len (y_train))

# Tamanho dos dados (amostras) para teste 
print('Dados de X para teste ', len(X_test))
print('Dados de y para teste ', len(y_test))
```

```python
# Importação das Bibliotecas

import pandas as pd # Importar dados externos

import numpy as np # Trabalhar com algebra linear

import matplotlib.pyplot as plt # Vizualiação de dados

from sklearn.linear_model import LinearRegression # Machine Learning, carregando o algoritmo de regressão
```

```python
#Carregando a base de dados

caminho = './Arquivos/cars.csv'

df = pd.read_csv(caminho)

df.shape

df.head()
```

```python
# Excluir a primeira coluna do dataframe

#Exclui a coluna ou linha, se Axis = 1 (Coluna) e Axis = 0 (linha)

df = df.drop(['Unnamed: 0'],axis=1) 

```

```python
df.head()
```

```python
# Separar as variaveis em independente (X) e dependente (y)

X = df.loc[:,'dist'].values
y = df.loc[:,'speed'].values

# Calculo de correlação entre variáveis
correlacao = np.corrcoef(X,y)

correlacao

# Formato de matriz com uma coluna a mais

X = X.reshape (-1,1)

# Criação do modelo e treinamento
# Fit indica que o treinamento deve ser executado

modelo = LinearRegression()
modelo.fit(X,y) # Treinar com esses dados
 
```

```python
#Visualização dos coeficientes
print('b=', modelo.intercept_)

#Inclinação
print ('a= ',modelo.coef_)
```

```python
# Prever a velocidade para uma distancia de 22 metros

modelo.predict([[22]])
```

```python
# Vizualizar os dados com gráfico de dispersão
plt.scatter (X,y)
plt.plot(X,modelo.predict(X), color = 'red')
```
