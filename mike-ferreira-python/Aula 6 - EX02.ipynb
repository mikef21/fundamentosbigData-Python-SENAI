```python
### i) 7. Calcule a soma, média, máximo, mínimo e contagem da coluna Preço.
#Soma
print(f'A soma resulta no valor de: {df["Preço"].sum()}')

#Média
print(f'O valor médio é: {df["Preço"].mean()}')

#Máximo
print(f'O maior valor é: {df["Preço"].max()}')

#Mínimo
print(f'O menor valor é: {df["Preço"].min()}')

#Contagem
print(f'Existem {df["Preço"].min()} itens na coluna preço')
```
