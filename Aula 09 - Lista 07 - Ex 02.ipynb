### Exercício 02 - Obtenha o lucro total de todos os meses e mostre o gráfico de linhas com as seguintes propriedades de estilo.

O gráfico de linha gerado deve incluir as seguintes propriedades de estilo:

• O estilo da linha pontilhada e a cor da linha devem ser vermelhas

• Mostrar legenda no local inferior direito.

• Nome do rótulo X = Número do mês

• Nome do rótulo Y = número de unidades vendidas

• Adicione um marcador de círculo.

• Cor do marcador de linha conforme lida

• A largura da linha deve ser 3



```python
# Importar as bibliotecas

import pandas as pd

import matplotlib.pyplot as plt

import os


# Mapear caminho do arquivo e alocar dadods

caminhoarquivo = './Arquivos/company_sales_data.csv'

df = pd.read_csv(caminhoarquivo)

# Cabeçalho

df.head()
```

```python
# Selecionar colunas

listameses= df['month_number'].to_list()
listalucro= df['total_profit'].to_list()

# Criar gráfico

plt.plot(listameses, # Eixo X
         listalucro, # Eixo Y
         label = " Lucro mensal do ano anterior", # Legenda
         marker = 'o', # Marcador
         c = 'red', # Cor da linha
         ls ='--', # Tipo de linha
         mec = 'black', # Cor do marcador
         linewidth = 4, # Espessura  linha
         markerfacecolor = 'k') # Remove a borda do marcador
         
# Adicionar titulos 

plt.xlabel ("Meses")
plt.ylabel ('Vendas em Dólares')
plt.title('Venda por mês do ano anterior')
plt.legend()

plt.yticks([100000,200000,300000,400000,500000]) # Ajustando os intervalores do eixo Y
plt.xticks(listameses) # Ajustando os gráficos do eixo X

plt.legend(loc = 'lower right')




```
