```python
# Importar as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = './arquivos/company_sales_data.csv'

# Alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

# Visulizar o cabeçalho do Dataframe
df.head()
```

```python
# Selecionar os dados que vão ser apresentados no gráfico
listaLucro = df['total_profit'].to_list()
listaMeses = df['month_number'].to_list()

# Criar o gráfico utilizandos os dados listados anteriormente
plt.plot(listaMeses, listaLucro, label = 'Lucro mensal do ano anterior')

plt.xlabel('Número do mês') # Título do eixo X
plt.ylabel('Lucro') # Título do eixo Y
plt.title('Lucro mensal da empresa') # Título do gráfico
plt.legend() # Legenda do gráfico
```

### Exercício 05: 

Leia os dados de vendas de cremes faciais e produtos de lavagem facial e mostre-os usando o gráfico de barras. O gráfico de barras deve exibir o número de unidades vendidas por mês para cada produto. Adicione uma barra separada para cada produto no mesmo gráfico. O gráfico de barras deve ficar assim.

```python
# Importar as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = './arquivos/company_sales_data.csv'

# Alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

# Visulizar o cabeçalho do Dataframe
df.head()

```

```python
# GERAR LISTAS
listaMeses = df['month_number'].to_list()
vendascreme = df['facecream'].to_list()
vendaslimpador = df['facewash'].to_list()

# GERAR GRAFICO DE BARRAS

plt.bar (listaMeses,vendascreme, label = 'Creme Dental')
plt.bar (listaMeses,vendaslimpador, label = 'Limpador Facial')

# TITULO
plt.title('GRÁFICO DE VENDAS')

#LEGENDA
plt.legend(loc='upper left', fontsize = 8)

```
