```python
# Importar as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = './arquivos/company_sales_data.csv'

# Alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

# Visulizar o cabe√ßalho do Dataframe
df.head()
```

```python
# Selecionar os dados que v√£o ser apresentados no gr√°fico
listaLucro = df['total_profit'].to_list()
listaMeses = df['month_number'].to_list()

# Criar o gr√°fico utilizandos os dados listados anteriormente
plt.plot(listaMeses, listaLucro, label = 'Lucro mensal do ano anterior')

plt.xlabel('N√∫mero do m√™s') # T√≠tulo do eixo X
plt.ylabel('Lucro') # T√≠tulo do eixo Y
plt.title('Lucro mensal da empresa') # T√≠tulo do gr√°fico
plt.legend() # Legenda do gr√°fico
```

### Exerc√≠cio 05: 

Leia os dados de vendas de cremes faciais e produtos de lavagem facial e mostre-os usando o gr√°fico de barras. O gr√°fico de barras deve exibir o n√∫mero de unidades vendidas por m√™s para cada produto. Adicione uma barra separada para cada produto no mesmo gr√°fico. O gr√°fico de barras deve ficar assim.

```python
# Importar as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = './arquivos/company_sales_data.csv'

# Alocar os dados em um dataframe
df = pd.read_csv(caminhoArquivo)

# Visulizar o cabe√ßalho do Dataframe
df.head()

```

```python
# GERAR LISTAS
listaMeses = df['month_number'].to_list()
vendascreme = df['facecream'].to_list()
vendaslimpador = df['facewash'].to_list()

# GERAR GRAFICO DE BARRAS

plt.bar (listaMeses,vendascreme, label = 'Creme Dental')
plt.bar (listaMeses,vendaslimpador, label = 'Limpador Facial')

# TITULO
plt.title('GR√ÅFICO DE VENDAS')

#LEGENDA
plt.legend(loc='upper left', fontsize = 8)

```
