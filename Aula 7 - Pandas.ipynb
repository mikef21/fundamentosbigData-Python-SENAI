### TRABALHANDO COM DATAFRAME



```python
## IMPORTARTAR BIBLIOTECA

import pandas as pd

## Caminho do arquivo

caminho = "C:\\Users\\FIC\\Documents\\mike-ferreira-python\\Arquivos\\bank-full.xlsx"

#Carregar um arquivo para um dataframe

df = pd.read_excel (caminho, sheet_name ='Sheet 1 - bank-full')

```

```python
#Visualizar as primeiras linhas do dataframe

df.head ()


```

```python
# Visualizar as ultimas linhas do dataframe

df.tail()
```

```python

# Modicar os nomes das colunas
# As colunas devem estar na mesma ordem dos nomes e a quantidade nomes deve ser igual a quantidade de colunas

df.columns = ['idade','job','estado civil', 'educação', 'default', 'balance', 'housing', 'empréstimo', 'contato', 'dia', 'mês', 'duração', 'campanha', 'pdays', 'previous', 'poutcome' ]

df.head()
```

```python
df = df.rename (columns= { 'idade':'IDADE', 'job':'Profissão'})

df.head()
```
