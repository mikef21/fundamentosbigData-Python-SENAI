Introdução ao Pandas II

Os dataframes do Pandas são estruturas muito versáteis por permitem uma série de operações que ajudam nas análises dos dados. Veremos agora as seguintes operações:

• Contagem de valores;

• Média;

• Soma;

• Valores únicos;

• Limpeza de dados duplicados;

• Limpeza de dados faltantes.

```python
# Importação da biblioteca

import pandas as pd


# Mapear Caminho do Arquivo
caminhoCompleto = './Arquivos/bank-full.xlsx'

# Alocar em um dataframe
df = pd.read_excel(caminhoCompleto, sheet_name = 'Sheet 1 - bank-full')

#Exibir o cabeçalho

df.head()
```

### Método value_counts

Não poucas vezes precisaremos saber a frequência com que uma dada informação se repete. Para obter tal informação de uma dada variável no Pandas podemos utilizar a função pd.value_counts().



```python
#Mostrar a quantidade de cada valor

df['education'].value_counts()

```

```python
# Calculando média em uma coluna

df.balance.mean()
```

```python
# Calculando mais de uma coluna de uma vez

df[['balance','duration']].mean()
```

### Unique e nunique

Outra importante função é a pd.unique(), que nos mostra todos os valores únicos em uma dada variável.

Já a função pd.nunique() nos mostra a quantidade de valores únicos em uma variável.


```python
# Listar elementos unicos de uma coluna
#Remove as duplicadas e retorna os elementos unicos

df.job.unique()

```

```python
#  Quantidade de valores únicos em uma coluna

df.job.nunique()
```

```python
 # Calcular a quantidade de registros duplicados

df.duplicated().sum()
```

```python
 # Calcular a quantidade de registros duplicados em uma coluna

 df.job.duplicated().sum()
```

```python
# Exibir quais linhas são duplicadas

df[df['job'].duplicated() == True]
```
