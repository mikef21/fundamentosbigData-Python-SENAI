```python
# a) Abra o arquivo World Happiness.

import pandas as pd

caminhoex06 = '../mike-ferreira-python/Arquivos/world_happiness_report_2015.csv'

df = pd.read_csv(caminhoex06)

df.head()
```

```python
# b) Qual a média do score_felicidade?

mediaFelicidade = df['happiness score'].mean()

print(f'A média é {mediaFelicidade: .2F}')
```

```python
# c) Qual a soma do PIB?

df = df.rename(columns  = {'economy (GDP per Capita)': 'PIB'})

df['PIB'].sum()
```

```python
# d) Qual a soma do freedom e corrupção?
df = df.rename(columns  = {'trust (Government Corruption)': 'corrupção'})


df[['freedom','corrupção']].sum()
```

```python
# e) Há dados duplicados? Quantos? Verifique quais são eles?

#Quantos valores?
print(f'Existem {df.duplicated().sum()} valores duplicados')

#Apresentar os itens 

print( f'')

print( f'Os valores duplicados são:')

df[df.duplicated()==True]

```

```python
# f) Verifique a quantidade de dados faltantes em cada coluna

print (f'A quantidade de dados faltantes está listada abaixo por coluna')
df.isna().sum()
```

```python
# g) Crie um dataframe onde os valores faltantes de score_felicidade sejam substituídos por -9999.

df = df.fillna(-9999)

df.isna().sum()
```
