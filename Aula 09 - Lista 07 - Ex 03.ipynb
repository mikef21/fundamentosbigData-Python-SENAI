### Exercício 03 – Leia todos os dados de vendas de produtos e mostre-os usando um gráfico multilinha.

Exiba o número de unidades vendidas por mês para cada produto usando gráficos multilinhas. (ou seja, faça a plotagem das linhas separadas para cada produto).



```python
# Importar as bibliotecas

import pandas as pd

import matplotlib.pyplot as plt

import os


# Mapear caminho do arquivo e alocar dadods

caminhoarquivo = './Arquivos/company_sales_data.csv'

df = pd.read_csv(caminhoarquivo)

# Cabeçalho

df.head()
```

```python
# Criar lista

listameses = df ['month_number'].to_list()

facecreamsalesdata = df ['facecream'].to_list()

facewashsalesdata = df ['facewash'].to_list()

facepastesalesdata = df ['toothpaste'].to_list()

bathingsoapsalesdata = df ['bathingsoap'].to_list()

shampoosalesdata = df ['shampoo'].to_list()

moisturizersalesdata = df ['moisturizer'].to_list()

#Criar gráfico

plt.plot (listameses,facecreamsalesdata,label = 'Face Cream', marker = 'o', linewidth = 3)
plt.plot (listameses,facewashsalesdata,label = 'Face Wash', marker = 'o', linewidth = 3)
plt.plot (listameses,facepastesalesdata,label = 'Face Paste', marker = 'o', linewidth = 3)
plt.plot (listameses,bathingsoapsalesdata,label = 'Bathing Soap', marker = 'o', linewidth = 3)
plt.plot (listameses,shampoosalesdata,label = 'Shampoo', marker = 'o', linewidth = 3)
plt.plot (listameses,moisturizersalesdata,label = 'Moisturizer', marker = 'o', linewidth = 3)

plt.yticks([1000,2000,3000,4000, 5000, 6000, 7000, 8000, 9000, 10000,11000,12000,13000,14000, 15000, 16000, 17000, 18000, 19000, 20000]) # Ajustando os intervalores do eixo Y
plt.xticks(listameses)

# Titulos dos eixos

plt.xlabel ('Meses')
plt.ylabel ('Vendas')

plt.title('Dados de Venda')

plt.legend(loc = 'upper left')
```
