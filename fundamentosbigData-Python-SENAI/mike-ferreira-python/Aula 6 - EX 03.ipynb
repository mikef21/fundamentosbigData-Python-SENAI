```python
## j) Crie um programa para criar um dataframe para mostrar as vendas por cargo, e salvar um arquivo em csv para com o nome de cada cargo:


## Filtrando dados estilo DIADEMA

## removendo itens duplicados na coluna cargos

df_cargos = df['Cargo'].drop_duplicates()

## Convertendo o df_cargos em lista

cargos = df_cargos.to_list()

## Filtrar utilizando a lista

for Cargo in cargos:
    df_final = df[df['Cargo'] == Cargo]  #Filtrar de acorco com os cargos
    caminho = './Arquivos/Relatórios EX.03/' # Caminho da pasta
    arquivo = Cargo +'.xlsx' # Nome do arquivo
    df_final.to_excel (caminho + arquivo) #Exportação do arquivo em excel

```

```python
## Lista Vendedor

## removendo itens duplicados na coluna cargos

df_vendedores = df['Vendedor'].drop_duplicates()

## Convertendo o df_cargos em lista

vendedores = df_vendedores.to_list()

## Filtrar utilizando a lista

for Vendedor in vendedores:
    df_final = df[df['Vendedor'] == Vendedor]  #Filtrar de acorco com os cargos
    caminho = './Arquivos/Relatórios EX.03/' # Caminho da pasta
    arquivo = Vendedor +'.xlsx' # Nome do arquivo
    df_final.to_excel (caminho + arquivo) #Exportação do arquivo em excel
```

```python
## Lista Clientes

## removendo itens duplicados na coluna cargos

df_clientes = df['Cliente'].drop_duplicates()

## Convertendo o df_cargos em lista

clientes = df_clientes.to_list()

## Filtrar utilizando a lista

for Cliente in clientes:
    df_final = df[df['Cliente'] == Cliente]  #Filtrar de acorco com os cargos
    caminho = './Arquivos/Relatórios EX.03/' # Caminho da pasta
    arquivo = Cliente +'.xlsx' # Nome do arquivo
    df_final.to_excel (caminho + arquivo) #Exportação do arquivo em excel
```

```python
## Lista UF

## removendo itens duplicados na coluna cargos

df_uf = df['UF'].drop_duplicates()

## Convertendo o df_cargos em lista

uf = df_uf.to_list()

## Filtrar utilizando a lista

for UF in uf:
    df_final = df[df['UF'] == UF]  #Filtrar de acorco com os cargos
    caminho = './Arquivos/Relatórios EX.03/' # Caminho da pasta
    arquivo = UF +'.xlsx' # Nome do arquivo
    df_final.to_excel (caminho + arquivo) #Exportação do arquivo em excel
```
