Você é o proprietário de uma sorveteria que opera em uma cidade ensolarada. Como proprietário, você deseja maximizar seus lucros e garantir que seu estoque de sorvetes seja gerenciado eficientemente. Para fazer isso, você decidiu criar um modelo de previsão de vendas de sorvetes com base na
temperatura diária. 
Foram coletados dados históricos de temperatura e receita em dólares ao longo de vários meses para criar um modelo de regressão linear. Seu objetivo é prever a receita diária em dólares com base na temperatura do ar externo (em graus Celsius) para que você possa tomar decisões informadas sobre o estoque e a produção de sorvetes.

### ETAPA 05: CRIAR E TESTAR O MODELO


```python
# Criação da variaveis dependente (Y)
y= df['Revenue']

# Criação da variaveis independente (X)
X = df[['Temperature']]

# Separar os dados para Teste e Treino 

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)

#Tamanho dos dados (amostras) de treino
print('Dados de X para treinamento', len (X_train))
print('Dados de y para treinamento', len (y_train))

# Tamanho dos dados (amostras) para teste 
print('Dados de X para teste ', len(X_test))
print('Dados de y para teste ', len(y_test))

```

```python
#Correlação 

correlacao = np.corrcoef (X,y)
correlacao
```

```python
# Criação do modelo

modelo = LinearRegression()

# Treinamento do modelo

modelo.fit(X_train, y_train)
```

```python
print('Coeficiente linear (b)',modelo.intercept_)
print('Coeficiente angular (a)',modelo.coef_)
```

```python
# ETAPA 07:  TESTANDO O MODELO

temperatura = 41

y_predict = modelo.predict(X_test)

previsao = modelo.predict([[temperatura]])

print(f'O seu faturamento com a temperatura de  {temperatura} °, será de {previsao} reais')
```
